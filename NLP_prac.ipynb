```python
text="Artificial Inteligence is smart and making humans work easier"

#printing tekns from text
tokens=word_tokenize(text)
print(tokens)
```

```python
#word frequency
word_freq=FreqDist(tokens)
print("Word Frequency is :",word_freq)
```

```python
# count word frequency
count=Counter(tokens)
print("Word frequency is:",count)
```

```python
#Removing the Stop words
stop_words=set(stopwords.words('english'))
filtered_words=[]

for word in tokens:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("Removed stop words then text is :",filtered_words)
```

```python
#POS tagging
pos_tagging=pos_tag(filtered_words)
print("POS Tagging are:",pos_tagging)
```
