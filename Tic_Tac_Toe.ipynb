```python
#Tic TAc toe game

#Creating the 9 empty spaces
board=[' ' for i in range(9)]

def print_Board():
    print()
    print(board[0],"|",board[1],"|",board[2])
    print("---------------")
    print(board[3],"|",board[4],"|",board[5])
    print("---------------")
    print(board[6],"|",board[7],"|",board[8])
    print("---------------")

# checing the winner
def check_winner(player_move):

    win_positions=[
        [0,1,2],
        [3,4,5],
        [6,7,8],

        [0,3,6],
        [1,4,7],
        [2,5,8],

        [0,4,8],
        [2,4,6]
    ]

    for pos in win_positions:
        if board[pos[0]]==board[pos[1]]==board[pos[2]]==player_move:
            return True
    return False

#main logic
current_player="X"

for turn in range(9):

    print_Board()

    move=int(input(f"Player:{current_player},Enter the position(0-8)"))

    # check if posiition is empty 
    if board[move]==' ':
        board[move]=current_player
    else:
        print("Position already taken !")
        continue

    #check winner 
    if check_winner(current_player):
        print_Board()

        print(f"Player{current_player} Wins !")
        break;

    #Change the player
    if current_player=="X":
        current_player="O"
    else:
        current_player="X"

#draw condition
else:
    print_Board()
    print("Game Draw !")
        
```
